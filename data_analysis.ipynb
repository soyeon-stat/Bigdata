#### 2. 작업 과정(Pseudo Code)

```python
dirty_content = "하나의 String으로 연결된 지저분한 상태의 원본 글"

# 1. 문장 나누기
sentence_list = 문장으로_나누기(dirty_content)

# 2. 문자열 전처리
cleaned_sentence_list = []
for sent in sentence_list :
    clean_sentence = 문자열_정리(sent)
    cleaned_sentence_list.append(clean_sentence)

# 3. 문장의 핵심어 찾기
keyword_list = []
for sent in cleaned_sentence_list :
    keyword = 핵심어_찾기(sent)
    keyword_list.append(keyword)

# 4. Keyword에 대한 감성분석
sentiment_df = pd.DataFrame()

for keyword, sent in zip(keyword_list, cleaned_sentence_list) :
    sentiment = 문장에서_키워드에_대한_감성분석(keyword, sent)
    tmp = pd.DataFrame({'keyword' : keyword, 'sentiment' : sentiment})
    sentiment_df = pd.concat([sentiment_df, tmp])

# 5. 키워드별 감성 분석 집계
result = sentiment_df.groupby('keyword')[['sentiment']].mean() # 키워드가 여러 번 출현하는 경우 점수를 평균

```

#### 3. 유의할 사항들?

- "특정 키워드"에 어떤 문장의 감성 분석을 하려면 문장 나누기를 잘 하면 될 것 같음
  - "특정 키워드"에 대해 감성분석을 하는 경우, 특정 키워드가 조건부로 주어져야 함
  - 다만 문장나누기가 충분히 잘 된다면 해당 문장의 키워드 추출과 해당 문장의 감성분석을 독립적으로 해도 무방할 것으로 보임
    * e.g.) "홍길동은 백성들을 구한 의적이고, 고을 원님은 욕심많은 사람이다"를 "홍길동은 백성들을 구한 의적", "고을 원님은 욕심많은 사람"으로 나눈다면 키워드-감성분석을 독립적으로 해도 무방
    * 만약 "홍길동은 백성들을 구한 의적이고, 고을 원님은 욕심많은 사람이다"가 문장나누기가 안된다면 바로 감성분석을 하는 경우 이상한 결과 도출될 수 있음
 
 
 → 의미역 분석(Semantic Role Labeling)을 잘 이용하면 가능할 것으로 보임
 → 관련 연구 있음 ([링크](https://www.dbpia.co.kr/journal/articleDetail?nodeId=NODE10487830))

## 활용할 만한 라이브러리 아카이빙

##### 문장나누기 라이브러리
- 대표적으로 `kss`와 `kiwipiepy`가 있음
- `kss`는 이름에 온점이 들어간 경우를 나누고
- `kiwipiepy`의 나누어야 할 문장을 잘 못 나눔

````python
# kss
from kss import split_sentences

text = """
H. G. 웰스부터 에드거 라이스 버로스, 레이 브래드버리, 필립 K. 딕 등에 이르기까지 미지의 행성 ‘화성’은 수많은 작가들에게 상상력의 원천이자 영감이 되어왔다. 그리고 오늘날 다시 한 번 화성에서 조난당한 한 남자에 대한 이야기로 전 세계적인 열풍을 일으키고 있는 작품이 있으니, 바로 이 책 《마션-어느 괴짜 과학자의 화성판 어드벤처 생존기》이다. 컴퓨터 프로그래머 출신인 작가 앤디 위어가 2009년 취미 삼아 개인 블로그에 연재를 시작했던 《마션》은 독자들의 요청으로 2011년 아마존 킨들 버전으로 자비 출판되었고, 이후 한 문학 에이전트의 눈에 띄어 미국의 중견 출판사 크라운 사에서 정식 출판되었다. “지난 수십 년을 통틀어 이토록 잘 읽히는 소설은 처음이다”, “21세기 과학적 지식이 빛을 발하는 스릴 넘치고 흥미진진한 이야기”라는 언론의 호평이 쏟아졌던 《마션》은 출간 즉시 뉴욕타임스 소설 부문 베스트셀러 순위권에 12주 연속 머물렀고, 1년도 채 안 된 오늘날 뉴욕타임스 소설 부문 베스트셀러 상위권에 37주 연속 재진입하고 있어 화제가 되고 있다. 이 책을 읽은 독자들의 폭발적인 관심에 힘입어 아마존 베스트셀러 TOP 10에 올랐으며, 보기 드물게 아마존 서평 12,894개, 별점 4.6에 이를 정도로 유례없는 인기를 자랑하고 있어 작가 및 작가 지망생 사이에서는 ‘퍼블리싱 드림’이라는 신조어가 생겨났을 정도다. 작년 이맘때만 해도 이런 상황을 전혀 예측하지 못했던 작가 앤디 위어는 여전히 믿기 어렵다는 반응을 보이며 “장기간에 걸친 신용 사기이거나, 누군가 나를 속이고 있다는 생각이 든다. 만약 그렇다면 최악의 상황이다. 누군가가 계속 내게 돈을 보내주고 있으니까”라 말해 사람들의 웃음을 자아내기도 했다.
"""

for sent in split_sentences(text) :
    print(sent)
````

````python
import urllib.request
from soynlp import DoublespaceLineCorpus
from soynlp.word import WordExtractor
from soynlp.tokenizer import LTokenizer
````

````python
# 학습에 사용한 문서 불러오기
urllib.request.urlretrieve("https://raw.githubusercontent.com/lovit/soynlp/master/tutorials/2016-10-20.txt", filename="2016-10-20.txt")
corpus = DoublespaceLineCorpus("2016-10-20.txt")
````

````python
# 해당 문서를 이용하여 학습 진행
word_extractor = WordExtractor()
word_extractor.train(corpus)
word_score_table = word_extractor.extract()
````

````python
# 2. 핵심어 추출하기
min_count = 2   # 단어의 최소 출현 빈도수 (그래프 생성 시)
max_length = 10 # 단어의 최대 길이
wordrank_extractor = KRWordRank(min_count=min_count, max_length=max_length)
beta = 0.85    # PageRank의 decaying factor beta
max_iter = 20
keywords, rank, graph = wordrank_extractor.extract(text_list, beta, max_iter)
````
