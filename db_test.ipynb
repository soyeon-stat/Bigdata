```python
from config.config import *
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine(f"postgresql+psycopg2://{DB_ID}:{DB_PW}@{DB_URL}:{DB_PORT}/{DB_NAME}")
```

```python
keyword = '계엄령'
query = f"""
SELECT 
    AVG(count_post) FILTER (WHERE count_post > 0) as avg_count_post
    ,AVG(count_like) FILTER (WHERE count_like > 0) as avg_count_like
    ,AVG(count_view) FILTER (WHERE count_view > 0) as avg_count_view
    ,AVG(count_comment) FILTER (WHERE count_comment > 0) as avg_count_comment
FROM dashboard
WHERE keyword = '{keyword}'
"""

data = pd.read_sql(query, engine)
data
```

```python
keyword = '계엄령'
query = f"""
SELECT community, post_date, count_post, count_like, count_view, count_comment
FROM dashboard
WHERE keyword = '{keyword}'
ORDER BY community, post_date
"""
data = pd.read_sql(query, engine)
data
```

```python
keyword = '계엄령'
query = f"""
SELECT community, COUNT(community) as CNT
FROM dashboard
WHERE keyword = '{keyword}'
GROUP BY community
"""
data = pd.read_sql(query, engine)
data
```

```python
keyword = '계엄령'
query = f"""
SELECT post_date, community, title, link, link_criterita, count_view, count_comment, count_like
FROM dashboard
WHERE keyword = '{keyword}'
and count_post > 0
ORDER BY count_post, count_comment desc
LIMIT 10
"""
data = pd.read_sql(query, engine)
data
```

```python
keyword = '코스피'
community = 'HUMORUIV'

query = f"""
SELECT community, post_date, label, label_score
FROM dashboard
WHERE keyword = '{keyword}'
and community = '{community}'
and count_post > 1
ORDER BY community, post_date
"""
data = pd.read_sql(query, engine)
data
```
