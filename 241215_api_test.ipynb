```python
import requests
import pandas as pd

domain = 'https://iqsclovpy6.execute-api.ap-northeast-2.amazonaws.com/'
api = 'best_post'
keyword = '계엄령'
start_date = '2024-12-01'
end_date = '2024-12-31'

url = f'{domain}/{api}?keyword={keyword}&start_date={start_date}&end_date={end_date}'
response = requests.get(url)
data = pd.read_json(response.json()['data'])
data
```
